```
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```
dataframe = pandas.read_csv("/content/model3_v1.csv")
dataframe
```

```
dataset = dataframe.values
dataset
```

```
X = dataset[:,1:].astype(float)
y = dataset[:,:1].astype(float)
```

```
#encode class values as integers
encoder = LabelEncoder()
encoder.fit(y) 
encoded_y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_y)

y1 = dummy_y
y1
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.3)
```

```
#model
model = Sequential()
model.add(Dense(40, input_dim=37, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(5, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=5)
```

```
model.save('model3.h5')
```

```
import tensorflow as tf

model = tf.keras.models.load_model('model3.h5')
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
open("converted_model3.tflite", "wb").write(tflite_model)
```
